```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
```

```python
#Загрузка данных - данные об удаленных раковых опухолей
df = pd.read_csv(r'D:\Khabarov\Репозиторий\MachineLearningCourse\DATA\cancer_tumor_data_features.csv')
df
```

```python
#Масштабирование
scaler = StandardScaler()
scaled_X = scaler.fit_transform(df)
```

```python
#Создание ковариационной матрицы
#Передаем масштабированные признаки и rowvar - False, чтобы считать ковариации только между признаками
covariance_matrix = np.cov(scaled_X,rowvar=False)
```

```python
eigen_values, eigen_vectors = np.linalg.eig(covariance_matrix)
eigen_vectors
```

```python
num_components = 2
 #Метод сортирует данные и возвращает их ИНДЕКСЫ
 #Сортируем их по убыванию через индексирование
 #Берем только первые 2
sorted_key = np.argsort(eigen_values)[::-1][:num_components]
sorted_key
```

```python
#Из исходных берем векторы и значения максимальные
eigen_values, eigen_vectors = eigen_values[sorted_key],eigen_vectors[:,sorted_key]
```

```python
#Проецируем исходные признаки на наши главные компоненты
principal_components = np.dot(scaled_X,eigen_vectors)
principal_components
```

```python
#Наносим на график
#plt.figure(figsize=(8,6),dpi=200)
plt.scatter(principal_components[:,0],principal_components[:,1])
```

```python
################### ЧЕРЕЗЕ SKLEARN ############################
scaler = StandardScaler()
scaled_X = scaler.fit_transform(df)
scaled_X
```

```python
from sklearn.decomposition import PCA

pca_model = PCA(n_components=2)

#Создаются главные компоненты
pca_model.fit(scaled_X)

#Проецируются в пространство главных компонент
pca_model.transform(scaled_X)
```

```python
#Комбинированный способ
pc_results = pca_model.fit_transform(scaled_X)
pc_results
```

```python
#Строим график
#У нас два признака, в X передаем первую колонку, в Y - вторую
plt.scatter(pc_results[:,0],pc_results[0:,1])
```

```python
#Компоненты - (2, 30)
pca_model.components_.shape
```

```python
#Какой признак какую долю вариативности ведет - [0.44272026, 0.18971182]
pca_model.explained_variance_ratio_
```

```python
#Какую долю вариативности данных несут наши новые признаки - 0.6324320765155942
np.sum(pca_model.explained_variance_ratio_)
```

```python
#Какую долю вариативности охватывают ВСЕ 30 признаков - 1.0
pca_30 = PCA(n_components=30)
pca_30.fit(scaled_X)
np.sum(pca_30.explained_variance_ratio_)
```

```python
#Анализ доли вариативности при уменьшении размерности
explained_variance = []

for n in range(1,30):
    pca = PCA(n_components=n)
    pca.fit(scaled_X)
    explained_variance.append(np.sum(pca.explained_variance_ratio_))
plt.plot(range(1,30),explained_variance)
plt.xlabel('Кол-во главных компонент')
plt.ylabel('Variance explained')
```
